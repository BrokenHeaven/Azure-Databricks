## Adding Azure DataStorage to Azure Databricks

Adding some instructions how to start working with Azure DataStorage

Using Scala to create connection to Azure Storage

```
%scala 
val containerName = "dbpystorecontainer"
val storageAccountName = "dbpystorage"
val sas = "?sv=2019-12-12&ss=bfqt&srt=sco&sp=rwdlacupx&se=2020-12-09T06:15:32Z&st=2020-12-08T22:15:32Z&spr=https&sig=S%2B0nzHXioi85aW%2FpBdtUdR9vd20SRKTzhNwNlcJJDqc%3D"
val config = "fs.azure.sas." + containerName+ "." + storageAccountName + ".blob.core.windows.net"
```

Using dbutils.fs.mount to mont the file to DBFS

Get Key, scope and configuration key.

```
dbutils.secrets.help()
```

Fire up Databricks CLI if and use following bash comannds for you to set the credentials.

```
%bash

databricks secrets create-scope --scope dbpyscope --scope-backend-type AZURE_KEYVAULT --resource-id xxxxx --dns-name 
```

```
%bash

databricks secrets create-scope --scope dbpyscope
databricks secrets create-scope --scope dbpyscope --initial-manage-principal "users"

databricks secrets put-acl --scope dbpyscope --principal "users"  --permission MANAGE
```

Once you have the configure keys generated, you are good to go.

```
dbutils.fs.mount(
  source = "wasbs://dbpystorecontainer@dbpystorage.blob.core.windows.net/Day9_MLBPlayers.csv",
  mount_point = "/Filestore/day9",
  extra_configs = {"conf-keyxxxx":dbutils.secrets.get(scope = "dbpyscope", key = "xxxx")})
```

And now start exploring the data :-)

```
%scala

val mydf1 = spark.read
.option("header","true")
.option("inferSchema", "true")
.csv("/mnt/storage1/Day9_MLBPlayers.csv")
display(mydf1)
```
